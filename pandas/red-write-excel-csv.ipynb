```python
import pandas as pd
df = pd.read_csv("stock.csv")
df
```

```python
df = pd.read_csv("stock.csv", skiprows=1) # or provide header=1
df
```

```python
df = pd.read_csv("stock.csv", header=None, names=[1, 2, 3, 4, 5])
df
```

```python
df = pd.read_csv("stock.csv",skiprows=1, nrows=3)
df
```

```python
df = pd.read_csv("stock.csv",skiprows=1, na_values=["not available", "n.a."])
df
```

```python
df = pd.read_csv("stock.csv",skiprows=1, na_values={"eps":["not available", "n.a."], "revenue":["not available", "n.a.", -1],
                                                   "people":["not available", "n.a."]})
df
```

```python
df.to_csv('new-stock.csv') # this creates a CSV with default index 0, 1, 2 ...
df.to_csv('new-stock.csv', index=False)
# df.to_csv('new-stock.csv', index=False, columns=['tickers', 'eps'])
# df.to_csv('new-stock.csv', index=False, header=False)
```
