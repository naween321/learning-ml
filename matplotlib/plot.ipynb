```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [50, 51, 52, 48, 47, 49, 45]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x, y, color='green', linewidth=5)
```

```python
plt.plot(x, y, 'rD')
```

```python
plt.plot(x, y, '--r*')
```

```python
plt.plot(x, y, 'rD--')
```

```python
days = [1, 2, 3, 4]
max_t = [50, 51, 52, 48]
min_t = [20, 21, 25, 18]
avg = [35, 38, 45, 33]
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(days, max_t, label="Max")
plt.plot(days, min_t, label="Min")
plt.plot(days, avg, label="Avg")
plt.legend(loc="best")
```
